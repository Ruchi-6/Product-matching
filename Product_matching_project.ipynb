```python
import os
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import nltk
```

```python
amz_data = pd.read_csv('amz_com-ecommerce_sample.csv' , encoding= 'unicode_escape')
amz_data.head()
```

```python
fk_data = pd.read_csv('flipkart_com-ecommerce_sample.csv' , encoding= 'unicode_escape')
fk_data.head()
```

```python
data = pd.concat([amz_data , fk_data] , axis = 0)
```

```python
data.head(10)
```

```python
data.columns
```

```python
data = data[['product_name' , 'product_category_tree' , 'pid' , 'retail_price' , 'discounted_price',
            'description']]
```

```python
data.duplicated().sum()
```

```python
data.isnull().sum()
```

```python
data = data.fillna(data.median())
```

```python
data.isnull().sum()
```

```python
data['description'] = data['description'].fillna("")
```

```python
data.head()
```

```python
data.columns
```

```python
data['product_category_tree']=data['product_category_tree'].map(lambda x:x.strip('[]'))
data['product_category_tree']=data['product_category_tree'].map(lambda x:x.strip('"'))
data['product_category_tree']=data['product_category_tree'].map(lambda x:x.split('>>'))
```

```python
data['retail_price'] = data['retail_price'].astype(str)
```

```python
data['discounted_price'] = data['discounted_price'].astype(str)
```

```python
data.head()
```

```python
data['product_category_tree'] = [','.join(map(str, l)) for l in data['product_category_tree']]
```

```python
data['Combined_description'] = data['product_category_tree']+data['description']
```

```python
data.head()
```

```python
new_df = data[['product_name' , 'Combined_description']]
```

```python
new_df.head()
```

```python
new_df['Combined_description'] = new_df['Combined_description'].apply(lambda x : x.lower())
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(lemmatizer.lemmatize(i))
        
    return " ".join(y)
```

```python
new_df['Combined_description'] = new_df['Combined_description'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 3000 , stop_words = 'english')
vectors = cv.fit_transform(new_df['Combined_description']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(vectors)
```

```python

def product_match(product):
    product_index = data[data['product_name'] == product].index[0]
    distance = similarity_score[product_index]
    matched_product = sorted(list(enumerate(distance)) , reverse = True , key = lambda x : x[1])[1:2]
    
    for i in matched_product:
        print(data.iloc[i[0]].product_name)
    
    
```
